```python
import numpy as np
import matplotlib.pyplot as plt
import skfuzzy as fuzz
```

```python
Domain = np.linspace(0, 1, 100)
Domain
```

```python
A = fuzz.membership.trapmf(Domain, (0, 0.4, 0.6, 1))

plt.grid()
plt.plot(A)
plt.title("A fuzzy Set")

plt.savefig("A fuzzy Set.png")

plt.show()
```

```python
B = fuzz.membership.trapmf(Domain, (0.1, 0.2, 0.5, 0.9))

plt.grid()
plt.plot(B)
plt.title("B fuzzy Set")

plt.savefig("B fuzzy Set.png")

plt.show()
```

```python
def Minimum_Zade(a, b):
    return np.minimum(a, b)

Tnorm_min_AB = Minimum_Zade(A, B)
plt.grid()
plt.plot(Tnorm_min_AB)
plt.title("Minimum Zade")

plt.savefig("TMinimum Zade.png")

plt.show()
```

```python
def Algebraic_product(a, b):
    return a * b

Tnorm_AP_AB = Algebraic_product(A, B)
plt.grid()
plt.plot(Tnorm_AP_AB)
plt.title("Algebraic Product")

plt.savefig("TAlgebraic Product.png")

plt.show()
```

```python
def Bounded_product(a, b):
    return np.maximum(a + b - 1, 0)

Tnorm_BP_AB = Bounded_product(A, B)
plt.grid()
plt.plot(Tnorm_BP_AB)
plt.title("Bounded Product")

plt.savefig("TBounded Product.png")

plt.show()
```

```python
def Drastic_Product(a, b):
    if b == 1:
        return a
    elif a == 1:
        return b
    else:
        return 0
    
vDrastic_Product = np.vectorize(Drastic_Product)
Tnorm_DP_AB = vDrastic_Product(A, B)
plt.grid()
plt.plot(Tnorm_DP_AB)
plt.title("Drastic Product")

plt.savefig("TDrastic Product.png")

plt.show()
```

```python
plt.subplot()

plt.plot(Tnorm_min_AB, label = "MIN")
plt.plot(Tnorm_AP_AB, label = "AP")
plt.plot(Tnorm_BP_AB, label = "BP")
plt.plot(Tnorm_DP_AB, label = "DP")

plt.grid()
plt.legend()

plt.savefig("Tnorm non-Parametric.png")

plt.show()
```

```python
def Maximum_Zade(a, b):
    return np.maximum(a, b)

Snorm_Max_AB = Maximum_Zade(A, B)
plt.grid()
plt.plot(Snorm_Max_AB)
plt.title("Maximum Zade")

plt.savefig("SMaximum Zade.png")

plt.show()
```

```python
def Algebraic_Sum(a, b):
    return a + b - (a * b)

Snorm_AS_AB = Algebraic_Sum(A, B)
plt.grid()
plt.plot(Snorm_AS_AB)
plt.title("Algebraic Sum")

plt.savefig("SAlgebraic Sum.png")

plt.show()
```

```python
def Bounded_Sum(a, b):
    return np.minimum(a + b, 1)

Snorm_BS_AB = Bounded_Sum(A, B)
plt.grid()
plt.plot(Snorm_BS_AB)
plt.title("Bounded Sum")

plt.savefig("SBounded Sum.png")

plt.show()
```

```python
def Drastic_Sum(a, b):
    if b == 0:
        return a
    elif a == 0:
        return b
    else:
        return 1
    
vDrastic_Sum = np.vectorize(Drastic_Sum)
Snorm_DS_AB = vDrastic_Sum(A, B)
plt.grid()
plt.plot(Snorm_DS_AB)
plt.title("Drastic Sum")

plt.savefig("SDrastic Sum.png")

plt.show()
```

```python
plt.subplot()

plt.plot(Snorm_Max_AB, label = "MAX")
plt.plot(Snorm_AS_AB, label = "AS")
plt.plot(Snorm_BS_AB, label = "BS")
plt.plot(Snorm_DS_AB, label = "DS")

plt.grid()
plt.legend()

plt.savefig("Snorm non-Parametric.png")

plt.show()
```
